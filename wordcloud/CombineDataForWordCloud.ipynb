```python
import pandas as pd
import os
import json
import glob
import csv
import re
import numpy as np
```

```python
all_data=[]
base_path='./type/json/'
path_list=os.listdir(base_path)
print(path_list)

path_list.sort(key=lambda x: int(x.split('.')[0]))

for fname in path_list:
    full_path = os.path.join(base_path, fname)
    print(fname)
    with open(full_path,'r',encoding='utf-8') as f:
        df = json.loads(f.read())
        for i in df:
            all_data.append(i)
all_data
```

```python
with open('./type/json/All_type.json', 'w', encoding="utf-8") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=4)
```
